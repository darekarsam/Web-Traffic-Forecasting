## 1. Data Exploration

```python
# Importing required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import numpy as np

%matplotlib inline
```

```python
# Reading Data
data = pd.read_csv("data.csv")
data.head(10)
```

### a. Features in the data

```python
data.dtypes
```

```python
#Convert date from string to datetime object
data.date = data.date.astype('datetime64[ns]')
```

```python
# check any missing values
data.isnull().any()
```

### b. Relation of car counts to cloud indicator variable

```python
sns.distplot(data[data['cloud.indicator']=='clear']['car.count'], hist=False, color='blue', label='Clear')
sns.distplot(data[data['cloud.indicator']=='cloudy']['car.count'], hist=False, color='grey', label='Cloudy')
plt.title('Distribution car count by cloud indicator')
plt.ylabel('Density')
```

There are 2 unique values for cloud indicator variable clear and cloudy, we can assume that some cloudy days are rainy and some are not, based upon general knowledge the number of cars should be more on clear days than those on rainy days as people prefer to stay indoors during the rain, but some cloudy days might not be rainy    

From the figure above we see that the distribution of car count variable for clear is on the higher side with less variance than that of the cloudy days(higher variance) as some cloudy days can be rainy which might explain the high variance in the data the higher end of the distribution might be for non rainy but cloudy days.      
Based on the above points we can say that the data seems reliable.    

### c. Dependence of #cars with time      

```python

fig, ax = plt.subplots(figsize=(20,10))
sns.lineplot(x='date', y='car.count', data=data)
ax.set_title("Time Series visualization")
```

From the above time series visualization we see that the car counts rose from 2010 to almost mid 2012 and then it started falling till 2016, due to abscence of future data we cannot confirm if this is a cyclic trend. The data is too dense to find more details of seasonality hence we will resample it later. I have decomposed the TS and plotted the rolling mean plot to confirm the trend.     

Below is the decomposition of the whole time series data showing the trend.

```python
res = sm.tsa.seasonal_decompose(data['car.count'], freq=300, model='additive')
resplot = res.plot()
```

```python
data[['year','month','day']] = data.date.apply(lambda x: pd.Series(x.strftime("%Y,%m,%d").split(",")))
# Set Date as index of Dataframe
indexed = data.set_index('date')
indexed.head()
```

```python
def rollingAnalysis(timeseries, windowsize=100):
    rolmean = timeseries.rolling(window=windowsize, center=False).mean()
    rolstd = timeseries.rolling(window=windowsize, center=False).std()
    return rolmean, rolstd

rolmean, rolstd = rollingAnalysis(indexed['car.count'])

fig, ax = plt.subplots(figsize=(20, 10), sharex=True)
indexed['car.count'].plot(color='green', label='Original')
rolmean.plot(color='red', label='100 days Rolling Mean')
rolstd.plot(color='black', label='100 days Rolling Std')
plt.legend(loc='best')
plt.ylabel('Car Counts')
plt.tight_layout()
plt.title('Visualizing trend using rolling Mean')
```

##### c1. Time series aggregation by Week       
There are too many data points to derive meaningful information hence lets aggregate the timeseries data by using mean value per week. The resample() function aggregates the TS and takes mean for the specified interval.      

```python
# Take mean for each week
indexedWeek = indexed.resample('W').mean() 
fig, ax = plt.subplots(figsize=(20,10))
plt.plot(indexedWeek.index, indexedWeek['car.count'])
ax.set_title('Resampled Data by Week')
ax.set_ylabel('mean of car count by week')
```

We see that aggregating it by week has fewer data points than the original series hence we can see a more clear picture, but still the seasonality is not evident, hence lets aggregate the TS by month.

##### c2. Time series aggregation by Month

```python
indexedMonth = indexed.resample('M').mean()
fig, ax = plt.subplots(figsize=(20,10))
plt.plot(indexedMonth.index, indexedMonth['car.count'])
ax.set_title('Resampled Data by Month')
ax.set_ylabel('mean of car count by month')
```

```python
res = sm.tsa.seasonal_decompose(indexedMonth['car.count'], freq=10, model='additive')
resplot = res.plot()
```

```python
indexedMonth1 = indexedMonth.reset_index()
indexedMonth1['year'] = indexedMonth1.date.dt.year
indexedMonth1['month'] = indexedMonth1.date.dt.month
fig, ax = plt.subplots(figsize=(15,10))
ax = sns.lineplot(x="month", y='car.count', hue='year', palette='Set1', legend='full', data=indexedMonth1)
plt.legend()
ax.set_title('Seasonality plot by Month')
```

After aggregating by month too we don't see any seasonality, so I breaked the data by year and month in the seasonality plot above. From the seasonality plot it becomes evident that there is no seasonality which repeats every year, there are some patterns like increase in month 9 than that of 8, but nothing conclusive.

```python
groupedByMonth = indexed.groupby(['month','year'])
groupedByMonthMean = groupedByMonth.mean()[['car.count']].reset_index()
pivotedMean = groupedByMonthMean.pivot(index='year', columns='month', values='car.count')
#zero mean normalization
# 1: substract mean
dfNorm = pivotedMean.sub(pivotedMean.mean(axis=1), axis=0)
# 2: divide by standard dev
dfNorm = dfNorm.div(pivotedMean.std(axis=1), axis=0)

fig, ax = plt.subplots(figsize=(15,7))
sns.heatmap(dfNorm, annot=True, linewidths=.5, cmap="coolwarm")
ax.set_title('Heatmap of Mean car count(Normalized by Year) by Month')
```

The heatmap by Year and Month shows that generally for month 6 and 9 the data is on higher side always as all the boxes are redish there is lot of variance in month 11 as we can see almost all colors in that column.

```python
groupedByDay = indexed.groupby(['day.of.week','year'])
groupedByDayMean = groupedByDay.mean()[['car.count']].reset_index()
pivotedMean = groupedByDayMean.pivot(index='year', columns='day.of.week', values='car.count')
#zero mean normalization
# 1: substract mean
dfNorm = pivotedMean.sub(pivotedMean.mean(axis=1), axis=0)
# 2: divide by standard dev
dfNorm = dfNorm.div(pivotedMean.std(axis=1), axis=0)
# Re-ordering columns
dfNorm = dfNorm[['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday','Saturday', 'Sunday']]

fig, ax = plt.subplots(figsize=(15,7))
sns.heatmap(dfNorm, annot=True, linewidths=.5, cmap="coolwarm")
ax.set_title('Heatmap of Mean car count(Normalized by Year) by Week Day')
```

The Heatmap by year and day of week shows that Saturday has higher number of cars for all the years except 2014, 2014 actually shows weird behaviour than other years as it has max cars on Tuesdays than other years which have lower cars.
